```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return x ** 2

def gradient(x):
    return 2 * x

# Gradient Descent
def gradient_descent(x_start, lr, epoch):
    x = x_start
    x_history = [x]
    for _ in range(epoch):
        dx = gradient(x)
        x -= lr * dx
        x_history.append(x)
        
    return x_history
```

```python
x_start = 20
lr = 0.3
epochs = 30

# Gradient Descent
x_history = gradient_descent(x_start, lr, epochs)
```

```python
# Visualization
x = np.linspace(-20, 20, 100)
y = func(x)

plt.figure(figsize=(10, 5))
plt.grid(True)
plt.plot(x, y, '-r', label='f(x) = x^2')
plt.scatter(x_history, func(np.array(x_history)), color='blue', marker='x', label='Gradient Descent Steps')
plt.title('Gradient Descent Visualization')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
